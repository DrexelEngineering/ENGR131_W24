## E. Create the engr131 Kernel

### Windows

0. Go to Anaconda Navigator Home

1. Look for the CMD.exe PROMPT application, and click Install beneath it.

2. Launch CMD.exe PROMPT application by clicking Launch beneath it.

3. In the command prompt window that opens, enter exactly the following at the prompts.

`conda activate engr131_24`

`conda install ipykernel`

`ipython kernel install --user --name=engr131`

`conda deactivate`

3. Then, close the command prompt window. Proceed to Step F. Test JupyterLab with engr131 Kernel.


### Mac or Linux

0. In your default browser, JupyterLab should be open.

1. Click on "Terminal" to open a terminal. 

2. In the terminal window, enter the following at the prompts.

`conda activate engr131_24`

`conda install ipykernel`

`ipython kernel install --user --name=engr131`

`conda deactivate`

3. Then, close the terminal window by click the "x" beside "Terminal 1".
